# autoViz

This is an example notebook to show the functionality of autoViz (EDA tool)

```python
#import libraries
import pandas as pd
import numpy as np
#!pip install autoviz
from autoviz.AutoViz_Class import AutoViz_Class
```

```python
file='crime.xlsx'
df=pd.read_excel(file)

df.head()
```

```python
AV = AutoViz_Class()
```

```python
filename = 'crime.xlsx'
sep = ""
dft = AV.AutoViz(
    filename,
    sep,
    'crimes_per_1mn',
    df,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
