# autoViz

This is an example notebook to show the functionality of autoViz (EDA tool)

```python
#import libraries
import pandas as pd
import numpy as np
#!pip install autoviz
from autoviz.AutoViz_Class import AutoViz_Class
```

```python
file='car_details.csv'
df=pd.read_csv(file)

df.head()
```

```python
AV = AutoViz_Class()
```

```python
filename = 'car_details.csv'
sep = ","
dft = AV.AutoViz(
    filename,
    sep,
    'selling_price',
    df,
    header=0,
    verbose=0,
    lowess=False,
    chart_format="svg",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
)
```
